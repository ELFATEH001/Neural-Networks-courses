```python
import torch
from torchvision.models import resnet18, ResNet18_Weights
import numpy as np
```

```python
model = resnet18(weights=ResNet18_Weights.DEFAULT)
data  = torch.rand((1, 3, 64, 64))
labels = torch.rand((1,1000))
```

```python
# predection (forward pass)
predection = model(data)
```

```python
# backward pass
loss = (predection - labels).sum()
loss.backward()
```

```python
print(loss.item())
```

```python
print(loss.item())
```

```python
optim = torch.optim.SGD(model.parameters(), lr=1e-2, momentum=0.9)
```

```python
optim.step()
optim.zero_grad()
```
