## Tensor attributes
you have three att: 
* shape,
* dtype,
* device

```python
tensor = torch.rand(3,4)
if torch.cuda.is_available():
    tensor = tensor.to('cuda')
    print("Device tensor is stored on: {}".format(tensor.device))
```

```python
tensor = torch.ones(4,4)
print(tensor)
tensor[:,0] =2
print(tensor)
```

```python
t1 = torch.cat((tensor, tensor, tensor), dim=1)
print(t1)
```

```python
ones = torch.ones((2,2))
twos = 2 * torch.ones((2,2))

print(ones)
print(twos)
```

```python
st = torch.stack((ones, twos),dim=2)
print(st.shape)
print(st)
```

```python
print(ones * twos)
print(ones @ twos)
```
